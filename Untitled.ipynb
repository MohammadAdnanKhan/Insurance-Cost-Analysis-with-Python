```python
max_price=df["price"].max()
```

```python
min_price=df["price"].min()
```

```python
bins=np.linspace(min_price,max_price,4)
group_names=["do kaudi ka","aukaat ke andar","aukaat se bahar"]
df["price-banned"]=pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
```

```python
bins=np.linspace(df["price"].min(),df["price"].max(),5)
group_names=["1 kaudi ka","2 kaudi ka","3 kaudi ka","4 kaudi ka"]
df["price-binned2"]=pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame 'df' has 'engine-size' and 'price' columns

sns.regplot(x="engine-size", y="price", data=df)  # Create the regression plot
plt.title("Price vs. Engine Size with Regression Line")  # Set the title
plt.ylim(0)  # Set the lower y-axis limit to 0 (no comma after 0)
plt.show()  # Display the plot
```
